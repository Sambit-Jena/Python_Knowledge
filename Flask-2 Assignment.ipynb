## Q1. Explain GET and POST methods.

GET and POST are two commonly used HTTP methods for communication between a web browser and a server. These methods define how data is sent and processed in an HTTP request.

#### GET METHOD:
In Python, GET method is commonly used to send HTTP GET requests to retrieve data from a server using URL parameter. The requests library in Python provides a convenient way to send HTTP requests, including GET requests, and handle the responses. The data sent using the GET method is visible in the URL, which means it can be bookmarked and shared. However, GET requests have limitations on the amount of data that can be sent, typically up to a few kilobytes.

#### POST METHOD:
Similarly the POST method is commonly used to send HTTP POST requests to submit data to a server. The requests library in Python provides a convenient way to send HTTP requests, including POST requests, and handle the responses.
The data sent using the POST method is not visible in the URL and is hidden in the request body, it's providing more security and privacy. POST requests can handle larger amounts of data compared to GET requests.

## Q2. Why is request used in Flask?

In Flask, the "request" object is used to handle incoming HTTP requests to get to our web application. It provides access to various attributes and methods that allow us to retrieve information about the request being made by the client.
The request object helps us to handle different HTTP methods like GET, POST, PUT, DELETE, etc and also we can check the request.method() attribute to determine the method used in the request and perform specific actions accordingly.

And also the request object allows us to extract dynamic parts of the URL called path variables or route parameters. The request object provides access to information about the client making the request.
There are many attributes like request.form, request.args, request.headers, etc. have been used to retrieve the various information.

## Q3. Why is redirect() used in Flask?

In Flask, the redirect() function is used to perform a redirect to a different URL within our web application. It allows us to send an HTTP redirect response to the client's browser, and instructing it to navigate to a new URL.
Bsically by  using the redirect() function in Flask, we can control the flow of our application, handle user navigation, and enforce best practices for URL management. It provides a convenient way to redirect users to different URLs within our web application.

Let's take some example
* When a user tries to access a protected resource but is not authenticated or authorized, we can use redirect() to redirect them to a login page or can show an access denied page.
* After processing a form submission, it is often a good practice to redirect the user to a different page rather than rendering the same page again. This helps the user to prevent duplicate entry.



## Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are files that contain HTML, along with placeholders for dynamic content, which can be filled in dynamically at  the runtime. Templates are used to separate the presentation logic from the application logic in web applications. They dynamically allow us to define the structure and layout of web pages and dynamically generate HTML based on data provided by our application.

#### Uses of render_template():
* The render_template() function is used to render a template file in Flask. It takes the name of the template file as a parameter and returns the rendered HTML content.
* It combines the template file with the data provided and can produces the final HTML output, which can then be sent as a response to the client's browser.
* The render_template() function also allows us to pass data from our created Flask routes or views to the template engine.
* Flask supports template inheritance, which allows us to define a base template that serves as a layout or structure for our web pages.
* By using the render_template() function in Flask, we can generate dynamic HTML pages by combining templates with data, separate our application logic from presentation, and achieve code reusability and maintainability.
