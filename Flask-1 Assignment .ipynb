## Q1. What is Flask Framework? What are the advantages of Flask Framework?


Flask is a micro web framework written in Python. It is designed to be lightweight, simple, and easy to use, making it a popular choice for building web applications and API's. This kind of flexibility allows developers to choose the tools and libraries they prefer.

##### Advantages of Flask Framework:

*  Flask allows developers to choose the components and tools what they actually want to use, and also giving them the freedom to structure their applications the way they prefer.
* Intentionally Flask leverages Python's simplicity and readability. It embraces Python's advantages and encourages developers to write clean and maintainable code which help them to build perfect web pages anf API's.
*  Flask is designed to be minimalistic, which means it has a small core and provides only the essential features needed for web development. Which helps developer to understand properly and write the code perfectly.
*  Flask is highly extensible and provides a wide range of extensions that can be added to enhance its functionality.
* If we talk about documentation Flask has excellent documentation that covers all aspects of the framework. And the created the documentation is clear, detailed, and includes many examples, making it easier for developers to learn and use Flask effectively.
* Flask provides built-in support for testing, allowing developers to write unit tests and integration tests for their applications. This makes it easier to ensure the correctness and robustness of the codebase.

## Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs or routes the website to the corresponding functions in our application. It allows us to define different routes for different parts of our application, and specifically when a user visits a specific URL, Flask knows which function to call to handle that request.

The app.route() decorator is used in Flask to define routes. It is a way to associate a URL with a Python function. 

App routing in Flask provides a clean and organized way to handle different URLs and their corresponding functionality. It allows us to separate the logic for different parts of our application and makes it easier to maintain and expand our codelines.

By using app routes(), we can create a RESTful API, build multiple web pages, handle different HTTP methods, and much more things according to our need.

And also app routing enables us to implement dynamic routes by including variable parts in the URL. These variables can be extracted and used in our functions to provide dynamic responses based on user input.

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function in Flask basically use to construct URLs. With the help of the route's endpoint, this function constructs a URL for a certain route. As it encapsulates the actual URLs and relies on the route endpoints, it helps to ensure that our application is flexible to changes in the URL format.

The url_for() function takes the endpoint name as its first argument and can also accept additional arguments to build the URL dynamically based on route parameters. It returns the URL as a string.
