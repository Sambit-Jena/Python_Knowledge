## Q1. What is a database? Differentiate between SQL and NoSQL databases.

A database is a structured collection of data that has been set up and saved to make it easy to manage and manipulate the information.

SQL and NoSQL are two different types of database management systems (DBMS) that are used to store and retrieve data, but they differ in their underlying principles and data models.

SQL Databases:

SQL (Structured Query Language) databases are based on the relational data model. They use a specified schema to specify the data's structure and the connections between various elements. Typically, SQL databases are set up as tables, which are made up of rows and columns. 

NoSQL Databases:

NoSQL databases are designed to handle large amounts of unstructured and semi-structured data. They do not impose a fixed schema and instead offer flexible data models that can adjust to changing requirements.  NoSQL databases are frequently utilised in situations where scalability, throughput, and high availability are essential.

## Q5. Explain Primary Key and Foreign Key.

In database management systems, a primary key and a foreign key are two different types of constraints that establish relationships between tables.

* Primary Key:
In a table, a primary key is a column or group of columns that uniquely identifies each record. Each table can have only one primary key and this cannot be contained any null values.t is typically created using a single column, but it can also be created using multiple columns.

* Foreign Key
A foreign key is a column or a set of columns in a table that refers to the primary key of another table. It establishes a link or a relationship between two tables based on the values in the foreign key columns. 
A table can have multiple foreign keys and it can be take references of the primary key of another table. The values in the foreign key column must either match an existing value in the referenced primary key column or it can be take null value.

## Q7. Give the order of execution of SQL clauses in an SQL query.

The order of execution of SQL clauses in an SQL query is as follows:

FROM: The FROM clause is the first clause to be executed in an SQL query. It specifies the table from which the data will be retrieved.

WHERE: Then the WHERE clause is executed after the FROM clause in SQL query. It's basically filters the rows based on the specified conditions.

GROUP BY: The GROUP BY clause is executed after the WHERE clause. This clause is used to groups the rows based on the specified columns.

HAVING: The HAVING clause may executed after the GROUP BY clause. It filterout the groups based on the specified conditions.

SELECT: The SELECT clause can be executed after any of the previous clauses. It specifies the columns to retrieve the data from the table.

ORDER BY: The ORDER BY clause can be executed just after the SELECT clause. It specifically sorts the result based on the specified columns by user.
